```python
#1.
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values is not None:
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have the same number of columns")
            self.values = [list(row) for row in values]
            self.n = len(values)
            self.m = len(values[0])
        elif n is not None and m is not None:
            if n <= 0 or m <= 0:
                raise ValueError("Matrix dimensions must be positive integers")
            self.values = [[0] * m for _ in range(n)]
            self.n = n
            self.m = m
        else:
            raise ValueError("You must provide either n and m, or a list of lists of values")

    def __getitem__(self, key):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            return self.values[i][j]
        elif isinstance(key, int):
            return self.values[key]
        else:
            raise IndexError("Invalid index. Use either M[i][j] or M[i, j]")

    def __setitem__(self, key, value):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            self.values[i][j] = value
        elif isinstance(key, int):
            if not isinstance(value, list) or len(value) != self.m:
                raise ValueError("Assigned row must be a list of the correct length")
            self.values[key] = value
        else:
            raise IndexError("Invalid index. Use either M[i][j] or M[i, j]")

    def __eq__(self, other):
        if isinstance(other, Matrix):
            return self.values == other.values
        return False

    def __repr__(self):
        return f"Matrix({self.values})"

    def assign(self, other):
        if isinstance(other, Matrix):
            if self.n != other.n or self.m != other.m:
                raise ValueError("Matrix dimensions must match for assignment")
            self.values = [list(row) for row in other.values]
        elif isinstance(other, list):
            if not all(len(row) == self.m for row in other) or len(other) != self.n:
                raise ValueError("The list of lists must match the dimensions of the matrix")
            self.values = [list(row) for row in other]
        else:
            raise ValueError("Assignment is only allowed with another matrix or a list of lists")

```

```python
#2.
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values is not None:
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have the same number of columns")
            self.values = [list(row) for row in values]
            self.n = len(values)
            self.m = len(values[0])
        elif n is not None and m is not None:
            if n <= 0 or m <= 0:
                raise ValueError("Matrix dimensions must be positive integers")
            self.values = [[0] * m for _ in range(n)]
            self.n = n
            self.m = m
        else:
            raise ValueError("You must provide either n and m, or a list of lists of values")

    def __getitem__(self, key):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            if isinstance(i, slice) or isinstance(j, slice):
                rows = self.values[i]
                return Matrix(values=[row[j] for row in rows])
            return self.values[i][j]
        elif isinstance(key, int):
            return self.values[key]
        else:
            raise IndexError("Invalid index. Use either M[i][j] or M[i, j]")

    def __setitem__(self, key, value):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            self.values[i][j] = value
        elif isinstance(key, int):
            if not isinstance(value, list) or len(value) != self.m:
                raise ValueError("Assigned row must be a list of the correct length")
            self.values[key] = value
        else:
            raise IndexError("Invalid index. Use either M[i][j] or M[i, j]")

    def __eq__(self, other):
        if isinstance(other, Matrix):
            return self.values == other.values
        return False

    def __repr__(self):
        return f"Matrix({self.values})"

    def assign(self, other):
        if isinstance(other, Matrix):
            if self.n != other.n or self.m != other.m:
                raise ValueError("Matrix dimensions must match for assignment")
            self.values = [list(row) for row in other.values]
        elif isinstance(other, list):
            if not all(len(row) == self.m for row in other) or len(other) != self.n:
                raise ValueError("The list of lists must match the dimensions of the matrix")
            self.values = [list(row) for row in other]
        else:
            raise ValueError("Assignment is only allowed with another matrix or a list of lists")

    def shape(self):
        return (self.n, self.m)

    def transpose(self):
        transposed_values = [[self.values[j][i] for j in range(self.n)] for i in range(self.m)]
        return Matrix(values=transposed_values)

    def row(self, n):
        if n < 0 or n >= self.n:
            raise IndexError("Row index out of bounds")
        return Matrix(values=[self.values[n]])

    def column(self, n):
        if n < 0 or n >= self.m:
            raise IndexError("Column index out of bounds")
        return Matrix(values=[[self.values[i][n]] for i in range(self.n)])

    def to_list(self):
        return [list(row) for row in self.values]

    def block(self, n_0, n_1, m_0, m_1):
        if not (0 <= n_0 < n_1 <= self.n) or not (0 <= m_0 < m_1 <= self.m):
            raise IndexError("Block indices out of bounds")
        block_values = [row[m_0:m_1] for row in self.values[n_0:n_1]]
        return Matrix(values=block_values)
```

```python
#3.
def constant(n, m, c):
    if n <= 0 or m <= 0:
        raise ValueError("Matrix dimensions must be positive integers")
    return Matrix(values=[[float(c) for _ in range(m)] for _ in range(n)])

def zeros(n, m):
    return constant(n, m, 0.0)

def ones(n, m):
    return constant(n, m, 1.0)

def eye(n):
    if n <= 0:
        raise ValueError("Matrix dimension must be a positive integer")
    return Matrix(values=[[1.0 if i == j else 0.0 for j in range(n)] for i in range(n)])


```

```python
#4.
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values is not None:
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have the same number of columns")
            self.values = [list(row) for row in values]
            self.n = len(values)
            self.m = len(values[0])
        elif n is not None and m is not None:
            if n <= 0 or m <= 0:
                raise ValueError("Matrix dimensions must be positive integers")
            self.values = [[0] * m for _ in range(n)]
            self.n = n
            self.m = m
        else:
            raise ValueError("You must provide either n and m, or a list of lists of values")

    def __getitem__(self, key):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            if isinstance(i, slice) or isinstance(j, slice):
                rows = self.values[i]
                return Matrix(values=[row[j] for row in rows])
            return self.values[i][j]
        elif isinstance(key, int):
            return self.values[key]
        else:
            raise IndexError("Invalid index. Use either M[i][j] or M[i, j]")

    def __setitem__(self, key, value):
        if isinstance(key, tuple) and len(key) == 2:
            i, j = key
            self.values[i][j] = value
        elif isinstance(key, int):
            if not isinstance(value, list) or len(value) != self.m:
                raise ValueError("Assigned row must be a list of the correct length")
            self.values[key] = value
        else:
            raise IndexError("Invalid index. Use either M[i][j] or M[i, j]")

    def __eq__(self, other):
        if isinstance(other, Matrix):
            return self.values == other.values
        return False

    def __repr__(self):
        return f"Matrix({self.values})"

    def assign(self, other):
        if isinstance(other, Matrix):
            if self.n != other.n or self.m != other.m:
                raise ValueError("Matrix dimensions must match for assignment")
            self.values = [list(row) for row in other.values]
        elif isinstance(other, list):
            if not all(len(row) == self.m for row in other) or len(other) != self.n:
                raise ValueError("The list of lists must match the dimensions of the matrix")
            self.values = [list(row) for row in other]
        else:
            raise ValueError("Assignment is only allowed with another matrix or a list of lists")

    def shape(self):
        return (self.n, self.m)

    def transpose(self):
        transposed_values = [[self.values[j][i] for j in range(self.n)] for i in range(self.m)]
        return Matrix(values=transposed_values)

    def row(self, n):
        if n < 0 or n >= self.n:
            raise IndexError("Row index out of bounds")
        return Matrix(values=[self.values[n]])

    def column(self, n):
        if n < 0 or n >= self.m:
            raise IndexError("Column index out of bounds")
        return Matrix(values=[[self.values[i][n]] for i in range(self.n)])

    def to_list(self):
        return [list(row) for row in self.values]

    def block(self, n_0, n_1, m_0, m_1):
        if not (0 <= n_0 < n_1 <= self.n) or not (0 <= m_0 < m_1 <= self.m):
            raise IndexError("Block indices out of bounds")
        block_values = [row[m_0:m_1] for row in self.values[n_0:n_1]]
        return Matrix(values=block_values)

    # New methods
    def scalarmul(self, c):
        return Matrix(values=[[c * self.values[i][j] for j in range(self.m)] for i in range(self.n)])

    def add(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix dimensions must match for addition")
        return Matrix(values=[[self.values[i][j] + N[i, j] for j in range(self.m)] for i in range(self.n)])

    def sub(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix dimensions must match for subtraction")
        return Matrix(values=[[self.values[i][j] - N[i, j] for j in range(self.m)] for i in range(self.n)])

    def mat_mult(self, N):
        if self.m != N.n:
            raise ValueError("Matrix dimensions must match for matrix multiplication")
        result_values = [[sum(self.values[i][k] * N[k, j] for k in range(self.m)) for j in range(N.m)] for i in range(self.n)]
        return Matrix(values=result_values)

    def element_mult(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrix dimensions must match for element-wise multiplication")
        return Matrix(values=[[self.values[i][j] * N[i, j] for j in range(self.m)] for i in range(self.n)])

    def equals(self, N):
        return self == N

```

```python
A = Matrix(values=[[1, 2], [3, 4]])
B = Matrix(values=[[5, 6], [7, 8]])
C = Matrix(values=[[1, 1], [1, 1]])
I = eye(2)

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix C:")
print(C)

print("\nIdentity Matrix I:")
print(I)

AB = A * B
(AB_C) = AB * C
BC = B * C
A_BC = A * BC

print("\nAssociative Property ((AB)C = A(BC)):")
print("(AB)C:")
print(AB_C)
print("A(BC):")
print(A_BC)
print("Are (AB)C and A(BC) equal?")
print(AB_C == A_BC)

B_plus_C = B + C
A_B_plus_C = A * B_plus_C
AB = A * B
AC = A * C
AB_plus_AC = AB + AC

print("\nDistributive Property (A(B + C) = AB + AC):")
print("A(B + C):")
print(A_B_plus_C)
print("AB + AC:")
print(AB_plus_AC)
print("Are A(B + C) and AB + AC equal?")
print(A_B_plus_C == AB_plus_AC)

BA = B * A

print("\nNon-Commutative Property (AB ≠ BA):")
print("AB:")
print(AB)
print("BA:")
print(BA)
print("Are AB and BA equal?")
print(AB == BA)

AI = A * I

print("\nIdentity Property (AI = A):")
print("AI:")
print(AI)
print("A:")
print(A)
print("Is AI equal to A?")
print(AI == A)
```
